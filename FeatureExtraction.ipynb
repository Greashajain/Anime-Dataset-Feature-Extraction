```python
#Function for Feature Extraction from Title in anime dataset
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True:
            data = data + i
        if i == '(':
            check = True
```

```python
#Applying the function to the column in dataset
df['Episodes']= df['Title'].apply(extract_episodes)
```

```python
#replacing the eps to ""
df['Episodes']= df['Episodes'].str.replace('eps',"")
```

```python
# the episode numbers are present in the String format , it is changed to integer
df['Episodes']=df['Episodes'].astype(int)
```

```python
df
```

```python
#extracting the time range from title
def Extraction_time(txt):
  check=False
  data=""
  for i in range(len(txt)):
    if txt[i]==')':
      for j in range(i+1,i+20):
        data +=txt[j]
      return data
```

```python
df['Total Time']= df['Title'].apply(Extraction_time)
```

```python
# Drop the unwanted column with None values if it exists (e.g., a wrongly named one)
df = df.loc[:, ~df.columns.duplicated()]  # removes duplicate columns
df = df.dropna(axis=1, how='all')         # drops columns that are entirely NaN or None
```

```python
print(df.columns)

```

```python
df.drop(columns=["total time"], inplace=True)
```

```python
print(df.columns)
```

```python
from dateutil.relativedelta import relativedelta
from datetime import datetime

def calculate_total_months(period):
    try:
        start_str, end_str = period.split(' - ')
        start_date = datetime.strptime(start_str, '%b %Y')
        end_date = datetime.strptime(end_str, '%b %Y')
        r = relativedelta(end_date, start_date)
        return r.years * 12 + r.months + 1  # +1 to include the starting month
    except:
        return None

df['Months'] = df['Total Time'].apply(calculate_total_months)
```
